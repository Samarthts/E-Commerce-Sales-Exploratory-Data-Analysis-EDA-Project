```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a sample DataFrame as 'sales_data.csv' was not found
data = {
    'price': np.random.uniform(10, 100, 100),
    'quantity_sold': np.random.randint(1, 50, 100),
    'revenue': np.random.uniform(100, 5000, 100),
    'rating': np.random.uniform(2.5, 5.0, 100),
    'category': np.random.choice(['Electronics', 'Clothing', 'Books', 'Home Goods'], 100),
    'region': np.random.choice(['North', 'South', 'East', 'West'], 100)
}
df = pd.DataFrame(data)

# 1. Dataset Overview
print(df.info())
print(df.head())

# 2. Data Quality Assessment
print("Missing values:\n", df.isnull().sum())
print("Duplicates:", df.duplicated().sum())
print("Statistics:\n", df.describe())

# 3. Univariate Analysis (Numeric)
for column in ['price', 'quantity_sold', 'revenue', 'rating']:
    plt.figure()
    sns.histplot(df[column].dropna(), kde=True)
    plt.title(f'Distribution of {column.capitalize()}')
    plt.show()

# 3. Univariate Analysis (Categorical)
for column in ['category', 'region']:
    plt.figure()
    df[column].value_counts().plot(kind='bar')
    plt.title(f'Distribution of {column.capitalize()}')
    plt.show()

# 4. Correlation Analysis
plt.figure(figsize=(7, 5))
corr = df[['price', 'quantity_sold', 'revenue', 'rating']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# 5. Key Insights
print("Highest price:", df['price'].max())
print("Best-selling category:", df.groupby('category')['quantity_sold'].sum().idxmax())
print("Region with highest revenue:", df.groupby('region')['revenue'].sum().idxmax())
```
